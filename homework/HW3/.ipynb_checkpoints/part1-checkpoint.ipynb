# Part 1: Classification


#### Training and Evaluating Models
Let's return back now to the problem of classification we dealt with in the previous tutorial but for now, we will bid the Philosophers goodbye, and focus instead on other forms of entertainment... Movie reviews, to be specific! The first step is to prepare the dataset for machine learning:

```python
#-------------------------------------------------------
# Import the rotton tomatoes dataset:
#-------------------------------------------------------
dataset   = load_dataset('rotten_tomatoes')

#-------------------------------------------------------
# Flatten out the dataset to get a single list of all the 
# sentences in the train, validation and test sets
# as well as the movie review labels.
#-------------------------------------------------------
sentences = dataset['train']['text']  + dataset['validation']['text'] + dataset['test']['text']
y         = dataset['train']['label'] + dataset['validation']['label'] + dataset['test']['label']

#-------------------------------------------------------
# Generate the bag-of-words representation of the text: 
#-------------------------------------------------------
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X          = vectorizer.fit_transform(sentences)

#-------------------------------------------------------
# Split the data into training (80%) and testing (20%) sets
#-------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, stratify = y, random_state = 123)
```

<br><br> The above code should be familiar to you from the last tutorial. We are simply taking the movie reviews data, creating a single python list of sentences (`X`), and a corresponding list of binary labels (`y`). As we did in the previous assignment, we can convert these sentences to their bag-of-words representation and, from there, try to predict the value of `y` using `X`. In the spirit of keeping with the previous assignment, let's try using a simple [logistic regression model](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) from `sklearn`: 

```python

from sklearn.linear_model    import LogisticRegression
from sklearn.model_selection import train_test_split,StratifiedKFold
from sklearn.metrics         import confusion_matrix

#-------------------------------------------------------
# Initialize four Logistic Regression models, 
# each with a different penalty
#-------------------------------------------------------
lr        = LogisticRegression(penalty  = 'none')                                                # No penalty
lr_l2     = LogisticRegression(penalty  = 'l2'        , solver = 'newton-cg' )                   # L2 penalty
lr_l1     = LogisticRegression(penalty  = 'l1'        , solver = 'saga'      )                   # L1 penalty
lr_l1l2   = LogisticRegression(penalty  = 'elasticnet', solver = 'saga' ,l1_ratio = 0.5)         # L1-L2 penalty

#-------------------------------------------------------
# Fit all four models using the training data
# Store a list of models for ease of evaluation later...
#-------------------------------------------------------
models = {}
models['Logistic_Regression']              = lr.fit(X_train,y_train)
models['Logistic_Regression_(L2)']         = lr_l2.fit(X_train,y_train)
models['Logistic_Regression_(L1)']         = lr_l1.fit(X_train,y_train)
models['Logistic_Regression_(Elasticnet)'] = lr_l1l2.fit(X_train,y_train)
```

<br><br>Let's compare the performance of the models on the training and the testing sets using the AUROC Curve:

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import auc

#-------------------------------------------------------
# Initialize a figure
#-------------------------------------------------------
fig, axs = plt.subplots(1,2, figsize=(16,8))

#-------------------------------------------------------
# Compute AUCs for Training and Testing Sets
#-------------------------------------------------------
for i,data_set in enumerate(['train','test']):
    

    if data_set == 'train':
        X = X_train; y = y_train
    else:
        X = X_test; y = y_test
        
    # For each of the model...
    for model in models.keys():

        # get the predicited probabilities on the test set
        probs    = models[model].predict_proba(X)

        # get the total set of the predictions for y = 1:
        thresholds = list(set(probs[:,1]))

        # for each threshold, transform the probabilities into predictions 
        # and get the true positive rate (tpr) as well as the false positive rate (fpr)
        tpr, fpr = [], []
        for t in thresholds:

            # Make predictions, given the threshold
            predict = [1 if x >= t else 0 for x in probs[:,1]]

            # Generate the element of the confusion matrix
            tp_i = sum([1 if (predict[i] == 1 and y[i] == 1) else 0 for i in range(len(predict))])   # True  Positives
            fp_i = sum([1 if (predict[i] == 1 and y[i] == 0) else 0 for i in range(len(predict))])   # False Positives
            fn_i = sum([1 if (predict[i] == 0 and y[i] == 1) else 0 for i in range(len(predict))])   # False Negatives
            tn_i = sum([1 if (predict[i] == 0 and y[i] == 0) else 0 for i in range(len(predict))])   # True  Negatives

            # compute the true positive rate and false positive rate
            tpr.append(tp_i / (tp_i + fn_i))
            fpr.append(fp_i / (fp_i + tn_i))

        # sort results by fpr for plotting purposes
        index = list(np.argsort(fpr))
        tpr,fpr = [tpr[i] for i in index], [fpr[i] for i in index] 

        # generate the label for the plot
        label = model.replace('_',' ') + '-- AUROC = ' + str(auc(fpr, tpr))[0:4]

        # Add the line to the plot
        axs[i].plot(fpr, tpr, label=label )
        
    # Decorate the plot
    axs[i].legend()
    axs[i].set_xlim((0, 1)) 
    axs[i].set_ylim((0, 1)) 
    axs[i].set_xlabel('False Positive Rate')
    axs[i].set_ylabel('True Positive Rate')
    axs[i].set_title(data_set + ' performance')
    
plt.tight_layout()
```

<br><br> Recall that for the AUROC measure, a value of 1 is the best you can get. With that in mind, this model is performing **excellently** on the training set; it's also performing pretty well on the testing set, especially when we consider that our bag-of-words representation is using simple uni-grams, and is consequently insensitive to the order of words! 

#### Regualrization and Hyperparamters

As we compare the difference in the models' performance across the training and the test sets, you may also notice that Logistic Regression  without regularization (blue line) has the best-in-class training set performance (AUC=0.99!) but those impressive results don't translate to the unseen testing set - where it has the worst-in-class performance. Anytime you observe large discrepancies between the testing and training performance, it is a red flag that your model has **over-fit** to the training data. And quite frankly, why wouldn't it? The standard logistic regression model's [loss function](https://ml-cheatsheet.readthedocs.io/en/latest/loss_functions.html) defines success as the minimization of error, exclusively; given our bag-of-words representation, the model has as many parameters to tune as the size of our vocabulary (i.e. a lot) and will take advantage of those many parameters to improve it's performance on the training set. In contrast, the logistic regression with the regularization term applies an additional cost as the coefficient values deviate from 0; intuitively, this makes the model consider the trade-off in the costs of tuning the parameter, and the consequent gains it obtains due to that tuning. 

In the figure above, it appears that the addition of the regularization term to the Logistic Regression's cost function improved the generalizability of the model; that is, regularization improved how closely the model's training set performance reflected its performance on tasks it had never seen before in the testing set. But one problem with our approach to regularization is that there is more than one way to apply it. Indeed, you probably noticed in the last homework that the `sklearn` implementation of elasticnet comes with a couple of `hyperparameters` (i.e. parameters that are not optimized by our model directly, but that we must select ourselves beforehand). When we perform `elasticnet` regularization we can control the strength of regularization (in sklearn, this is the `C` paramter), as well as the relative weight of the L1 and L2 penalties (in sklearn, this is controlled by the `l1_ratio` parameter). To help us understand the effects of these hyper-parameters on the performance of our models, let's train it with various settings of those hyper-parameters and see how it impacts the final AUROC:

```python
import os
import json 
from sklearn.metrics import roc_auc_score

#-------------------------------------------------------
# Check if the results are already computed, and if so, import them
#-------------------------------------------------------
if os.path.isfile('./materials/performance/logisticregression_aucs.json'):
    print('---------------------------------')
    print('       Importing Results         ')
    print('---------------------------------')
    with open('./materials/performance/logisticregression_aucs.json') as f:
        auc_c = json.load(f)

#-------------------------------------------------------
# Otherwise, compute the results and save them to disk.
#-------------------------------------------------------
else:    
    print('---------------------------------')
    print('       Generating Results        ')
    print('---------------------------------')
    
    auc_c = []
    # For various penality amplitudes
    for c in range(1,20,1):
        
        auc_r = []
    
        # For various L1-L2 penalty ratios.
        for r in range(0,20,1):

            #initialize the model
            lr   = LogisticRegression(penalty  = 'elasticnet', solver = 'saga', C = c/20, l1_ratio = r/20)   # L1-L2 penalty

            # fit the model
            lr.fit(X_train,y_train)

            # generate the prediction probablities
            probs  = lr.predict_proba(X_test)

            # get the AUC
            auc_r.append(roc_auc_score(y_test, probs[:,1]))

        auc_c.append(auc_r)
        
    # Save the results to disk  
    x = json.dumps(auc_c)
    f = open("materials/performance/logisticregression_aucs.json" , "w")
    f.write(x)
    f.close()
```

Now let's visualize the relationship between the model's performance, and the settings of the hyper-parameters `C` and `l1ratio` using the `heatmap` function in the [seaborn library](https://seaborn.pydata.org/):

```python
import seaborn as sns

#-------------------------------------------------------
#Generate the heatmap
#-------------------------------------------------------
plt.figure(figsize = (10,8))
ax = sns.heatmap(auc_c, cbar_kws={'label': 'AUROC (Test Set)'})

#-------------------------------------------------------
# Set the X Label and Tickmarks:
#-------------------------------------------------------
plt.xlabel("L1-L2 Ratio ('l1_ratio')")
plt.xticks(ticks   = range(0,20), 
          labels   = [str(0.05 * x)[0:4] for x in range(0,20)],
          rotation = 'vertical')

#-------------------------------------------------------
# Set the Y Labels and Tickmarks:
#-------------------------------------------------------
plt.ylabel("Regularization Amplitude ('C')")
plt.yticks(ticks    = range(0,20), 
           labels   = [str(0.05 * x)[0:4] for x in range(0,20)],
           rotation = 'horizontal')

plt.title("Logistic Regression performance for various hyperparamter settings")
```

<br><br> Notice that the settings of the hyper-parameters (`C` and `l1-ratio`) have tremendous consequences for the the model's performance on the unseen test set! If we set `C` to 0, and the `l1_ratio` to 1 (top right corner), our performance is much less than when we set the `l1_ratio` to 0 and `C` to 1 (bottom left). I bring this to your attention because, in-practice, the difference between a good and a great model may be due to the settings of the `hyper-paramters`, so make sure you explore them! 

#### A Smaller bag-of-words
As I mentioned earlier, one of the things that contributed towards the overfit of the logistic regression model without regularization was the abundance of `features` that get created when we generate a bag-of-words representation: 



```python
print('-------------------------------------------')
print('Size of the Rotton Tomatoes Training Data')
print('-------------------------------------------')
print(X_train.shape[0], 'x', X_train.shape[1])
```

<br><br> We have a whooping 18,330 unigrams in our dataset; that's over twice the number of rows that we have in our dataset! But, as you've probably come to suspect by now, not all unigrams are created equal. Let's start by counting the frequency of the unigrams in the text:

```python
#-------------------------------------------------------------
# Get the counts:
#-------------------------------------------------------------
counts = []
for i in range(0,X_train.shape[1]):
    counts.append(np.sum(X_train[:,i]))

#-------------------------------------------------------------
# Print the counts:
#-------------------------------------------------------------
for i in range(1,10):
    print(str(100*utils.CountFrequency(counts)[i] / X_train.shape[1])[0:4], '% of unigrams occur', i, 'times')
```

<br><br> Zipf's Law strikes again! It seems that 42.4% of the unigrams occur only once, 14.5% occur twice and so on. Let's see how our basic Logistic Regression model performs as we exclude/include unigrams from the bag-of-words according to their rank:

```python
import warnings
warnings.filterwarnings("ignore")

#-------------------------------------------------------------
# Initialize the plot
#-------------------------------------------------------------
fig, axs = plt.subplots(1,2, figsize=(16,6))

#-------------------------------------------------------------
# Plot the train performance of logistic regression given features
#-------------------------------------------------------------
aucs = []
features = []
for i,c in enumerate(set(counts)): 
    
    # Get the features we want to keep
    columns_to_keep = [i for i,x in enumerate(counts) if (x >= c)]
    features.append(X_train[:,columns_to_keep].shape[1])
    
    #fit the model using the selected features
    lr  = LogisticRegression(penalty  = 'none') 
    lr.fit(X_train[:,columns_to_keep],y_train)
    
    #compute the AUC and append
    probs = lr.predict_proba(X_train[:,columns_to_keep])
    aucs.append(roc_auc_score(y_train, probs[:,1]))
    if i == 100:
        break

# Sort and Plot        
index = list(np.argsort(features))
aucs,features = [aucs[i] for i in index], [features[i] for i in index] 
axs[0].plot(features, aucs)
axs[0].set_xlabel('Bag-of-Words Features Used (by Rank)')
axs[0].set_ylabel('True Positive Rate')
axs[0].set_ylim((0.6, 1))
axs[0].set_xlim((0, 5000))
axs[0].set_title('Training performance')
        
#-------------------------------------------------------------
# Plot the test performance of logistic regression given features
#-------------------------------------------------------------        
aucs = []
features = []
for i,c in enumerate(set(counts)): 
    
    # Get the features we want to keep
    columns_to_keep = [i for i,x in enumerate(counts) if (x >= c)]
    features.append(X_train[:,columns_to_keep].shape[1])
    
    #fit the model using the selected features
    lr.fit(X_train[:,columns_to_keep],y_train)
    
    #compute the AUC and append
    probs = lr.predict_proba(X_test[:,columns_to_keep])
    aucs.append(roc_auc_score(y_test, probs[:,1]))
    if i == 100:
        break

# Sort and Plot
index = list(np.argsort(features))
aucs,features = [aucs[i] for i in index], [features[i] for i in index] 
axs[1].plot(features, aucs)
axs[1].set_xlabel('Bag-of-Words Features Used (by Rank)')
axs[1].set_ylabel('AUROC')
axs[1].set_ylim((0.6, 1))
axs[1].set_xlim((0, 5000))
axs[1].set_title('Testing performance')
```

This above plot illustrates that we can toss out all but the first 1,600 or so words by rank (and words with a frequency of 12 or more), and increase the performance of the standard logistic regression model on the test set to 0.8, even without the regularization!

#### Even smaller bags-of-words with PCA:

The whole point of the regularization from the last section was to automatically identify features that aren't worth assigning a weight to during our model's optimization. The reason that regularization has such a positive impact is because a bag-of-words representation can become onerously large due to the realities of Zipf's law. To make matters worse, it is often (but not always) the case that the terms which occur infrequently provide little-to-no additional value for classification problems; we saw a clear example of this in the last section, where we were able to maximize the performance of our simple Logistic Regression model by reducing the 18,330 bag-of-words to a mere 1,600 of the most frequently used terms. Reducing ~90% of the unigrams (with no cost to our model's performance) is fantastic because, as I'm sure you appreciate at this point, even excellent personal computers struggle with performing computations on large arrays.  

But there is more compression to be done. [Principal Component Analysis](https://setosa.io/ev/principal-component-analysis/) is a way to cast our data onto a new set of orthogonal [basis vectors](https://en.wikipedia.org/wiki/Basis_(linear_algebra)) that are rank ordered according to the amount of variance they explain in the data: the first principal component explains the greatest amount of variance, followed by the 2nd principal component, and so on. Because of this rank ordering, we can take a subset of the principal components (say, the first $n$) that capture most of the variance in the original data, but with less columns to keep track of! 

Let's transform our data using PCA, and see if we can condense the size of our word vectors down any further without much cost to the model's performance:

```python
from sklearn.decomposition import PCA

#-------------------------------------------------------------
# Figure out which columns to keep based on the count of the occurances being 12 or less:
#-------------------------------------------------------------
c = 12
columns_to_keep = [i for i,x in enumerate(counts) if (x >= c)]

#-------------------------------------------------------------
# Perform PCA on the data
#-------------------------------------------------------------
pca = PCA()
pca.fit( X_train[:,columns_to_keep].toarray() )
X_train_pca = pca.transform(X_train[:,columns_to_keep].toarray())
X_test_pca  = pca.transform(X_test[:,columns_to_keep].toarray())
```

<br><br> We now have a transformed version of our original data in `X_train_pca` and `X_test_pca`. If we want to compress this data, we'll still need to choose how many principal components we want to keep. Let's plot the cumulative sum of the variance that is accounted for as a function of the principal components, and choose a number that will give us 90% of the original variance:

```python
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(0,len(columns_to_keep)),cumulative_variance)
plt.plot([0,1600],[0.9,0.9], '--', label='90% Variance')
plt.plot([700,700],[0.1,1.0],'--', label='First 750 Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('% of total variance accounted for')
plt.legend()
```

<br><br>It looks like we can retain 90% of the variance in our data using just 750 dimensions; that's less than half of the 1,600 we had earlier! Let's see how well our model performs with those 750 Principal components as the new features:

```python
n = 750
#-------------------------------------------------------------
#fit the model using only half the components (750 / 1600)
#-------------------------------------------------------------
lr  = LogisticRegression(penalty  = 'none') 
lr.fit(X_train_pca[:,0:n],y_train)

#-------------------------------------------------------------
#Compute the AUC
#-------------------------------------------------------------
probs = lr.predict_proba(X_test_pca[:,0:n])
aucs  = round(roc_auc_score(y_test, probs[:,1]),2)

print('-------------------------------------------------------')
print('AUC of the Logistic Regression Model using First 750 PCs')
print('-------------------------------------------------------')
print(aucs)
```

<br><br>And there you have it; with our simple exclusion and PCA, we have reduced the original bag-of-words representation from over 18 thousand, to a mere 750, at no cost to the model's performance.
